```python
"""
df_registry
==============
Lightweight DataFrame provenance tracker for Jupyter notebooks.

Tracks the lineage of pandas DataFrames as they are created and
transformed throughout a notebook pipeline.  Each call to
``register()`` appends one row to an in-memory registry that records:

    - the new DataFrame name and shape
    - the parent DataFrame name and shape
    - the operation label
    - any options / parameters passed to the transformation
    - a UTC timestamp

Typical usage
-------------
    from df_registry import register, get_registry, show_lineage, reset_registry

    # After every transformation, call register():
    df_clean = df_raw.drop(columns=["transaction_id", "card_id"])
    register(
        new_df=df_clean,
        new_name="df_clean",
        parent_name="df_raw",
        operation="drop_id_cols",
        options={"columns": ["transaction_id", "card_id"]},
    )

    # Display the full registry at any point:
    show_lineage()

Compliance
----------
Formatted with black and passes ruff (no bare `except`, no unused imports,
type annotations throughout).
"""

from __future__ import annotations

import json
from datetime import datetime, timezone
from typing import Any

# ---------------------------------------------------------------------------
# Module-level singleton registry
# ---------------------------------------------------------------------------

# Each element is a plain dict; converted to a DataFrame only on demand so
# that appending stays O(1) rather than O(n) per registration call.
_REGISTRY: list[dict[str, Any]] = []


# ---------------------------------------------------------------------------
# Public API
# ---------------------------------------------------------------------------


def register(
    *,
    new_df: pd.DataFrame,
    new_name: str,
    parent_name: str,
    operation: str,
    options: dict[str, Any] | None = None,
) -> None:
    """Record a DataFrame transformation in the provenance registry.

    Parameters
    ----------
    new_df:
        The newly created DataFrame (used to capture its shape).
    new_name:
        Variable name assigned to *new_df* in the notebook.
    parent_name:
        Variable name of the DataFrame that *new_df* was derived from.
    operation:
        Short label describing the transformation, e.g.
        ``"drop_id_cols"``, ``"encode_categoricals"``,
        ``"train_test_split"``.
    options:
        Arbitrary key/value pairs describing the parameters passed to
        the transformation (columns dropped, encoding method, split
        ratio, etc.).  Stored as a JSON string for compact display.

    Returns
    -------
    None
        Side-effect only: appends one row to the module-level registry.

    Examples
    --------
    >>> register(
    ...     new_df=df_encoded,
    ...     new_name="df_encoded",
    ...     parent_name="df_clean",
    ...     operation="encode_categoricals",
    ...     options={"cols": ["channel", "entry_mode"], "method": "onehot"},
    ... )
    """
    # Resolve parent shape from the registry if available so callers do not
    # have to pass the parent DataFrame object itself.
    parent_shape = _lookup_shape(parent_name)

    _REGISTRY.append(
        {
            # Lineage identifiers
            "new_name": new_name,
            "parent_name": parent_name,
            # Shapes — (rows, cols) tuples stored as strings for readability
            "new_shape": str(new_df.shape),
            "parent_shape": str(parent_shape) if parent_shape else "unknown",
            # Row / column deltas (None when parent shape is unknown)
            "row_delta": (
                new_df.shape[0] - parent_shape[0] if parent_shape is not None else None
            ),
            "col_delta": (
                new_df.shape[1] - parent_shape[1] if parent_shape is not None else None
            ),
            # Transformation metadata
            "operation": operation,
            # Serialise options to a compact JSON string; fall back to repr
            # for non-serialisable objects (e.g. sklearn transformers).
            "options": _safe_json(options or {}),
            # Wall-clock timestamp in UTC ISO-8601 format
            "registered_at": datetime.now(tz=timezone.utc).isoformat(
                timespec="seconds"
            ),
        }
    )


def get_registry() -> pd.DataFrame:
    """Return the full provenance registry as a pandas DataFrame.

    Returns
    -------
    pd.DataFrame
        One row per ``register()`` call, columns as described in the
        module docstring.  Returns an empty DataFrame (with the correct
        columns) if nothing has been registered yet.
    """
    columns = [
        "new_name",
        "parent_name",
        "new_shape",
        "parent_shape",
        "row_delta",
        "col_delta",
        "operation",
        "options",
        "registered_at",
    ]

    if not _REGISTRY:
        return pd.DataFrame(columns=columns)

    return pd.DataFrame(_REGISTRY, columns=columns)


def show_lineage(
    *,
    compact: bool = False,
    name_filter: str | None = None,
) -> pd.DataFrame:
    """Display and return the provenance registry.

    Intended for interactive use in Jupyter; the returned DataFrame is
    also displayed via ``pd.options.display`` settings so long lines are
    not truncated.

    Parameters
    ----------
    compact:
        If ``True``, hide the ``options`` and ``registered_at`` columns
        to fit the output on screen.
    name_filter:
        If provided, return only rows where ``new_name`` or
        ``parent_name`` contains this substring (case-insensitive).

    Returns
    -------
    pd.DataFrame
        The (possibly filtered / narrowed) registry view.
    """
    registry_df = get_registry()

    if registry_df.empty:
        print("Registry is empty — call register() after each transformation.")
        return registry_df

    # Optional name filter
    if name_filter is not None:
        mask = registry_df["new_name"].str.contains(
            name_filter, case=False, na=False
        ) | registry_df["parent_name"].str.contains(name_filter, case=False, na=False)
        registry_df = registry_df.loc[mask].reset_index(drop=True)

    # Optional column reduction for compact display
    if compact:
        display_cols = [
            "new_name",
            "parent_name",
            "new_shape",
            "row_delta",
            "col_delta",
            "operation",
        ]
        registry_df = registry_df[display_cols]

    # Widen display so options strings are not truncated
    with pd.option_context(
        "display.max_colwidth",
        120,
        "display.max_rows",
        100,
    ):
        # ``display()`` is available in Jupyter; fall back to print elsewhere.
        try:
            from IPython.display import display  # type: ignore[import-untyped]

            display(registry_df)
        except ImportError:
            print(registry_df.to_string(index=False))

    return registry_df


def reset_registry() -> None:
    """Clear all entries from the provenance registry.

    Useful at the top of a notebook to ensure a clean state on re-runs.

    Returns
    -------
    None
    """
    _REGISTRY.clear()
    print("Registry cleared.")


# ---------------------------------------------------------------------------
# Internal helpers
# ---------------------------------------------------------------------------


def _lookup_shape(name: str) -> tuple[int, int] | None:
    """Return the recorded shape of the most recent entry for *name*.

    Searches the registry in reverse order so the most recent
    registration for a given DataFrame name is found first.

    Parameters
    ----------
    name:
        The ``new_name`` to look up.

    Returns
    -------
    tuple[int, int] or None
        ``(rows, cols)`` if found, ``None`` otherwise.
    """
    for entry in reversed(_REGISTRY):
        if entry["new_name"] == name:
            # Shape is stored as a string like "(60000, 32)"; parse it back.
            shape_str: str = entry["new_shape"]
            try:
                rows, cols = (int(x.strip()) for x in shape_str.strip("()").split(","))
                return (rows, cols)
            except ValueError:
                return None
    return None


def _safe_json(obj: dict[str, Any]) -> str:
    """Serialise *obj* to a compact JSON string.

    Falls back to ``repr()`` for values that are not JSON-serialisable
    (e.g. numpy arrays, sklearn objects).

    Parameters
    ----------
    obj:
        Dictionary of options to serialise.

    Returns
    -------
    str
        JSON string, or ``repr(obj)`` on serialisation failure.
    """
    try:
        return json.dumps(obj, default=str, separators=(",", ":"))
    except (TypeError, ValueError):
        return repr(obj)
```

```python
def missing_report(df, sort_by="missing", ascending=False, pct_decimals=2):
    """Build a missingness summary for every column in *df*.

    Parameters
    ----------
    df : pd.DataFrame
        The DataFrame to inspect.
    sort_by : str, optional
        Column to sort results by.  ``"missing"`` sorts by raw count,
        ``"pct"`` by percentage.  Any other value disables sorting.
    ascending : bool, optional
        Sort direction.  ``False`` (default) puts the most-missing
        columns first.
    pct_decimals : int, optional
        Number of decimal places for the ``Sparsity %`` column.

    Returns
    -------
    pd.DataFrame
        One row per column with columns: ``Missing``, ``Sparsity %``,
        ``dtype``, and ``Example`` (first non-null value).
    """
    n_rows = len(df)

    # Build the summary frame — one row per column
    report = pd.DataFrame(
        {
            # Raw count of NaN / None values
            "Missing": df.isna().sum(),
            # Fraction of rows that are missing, expressed as a percentage
            "Sparsity %": (df.isna().sum() / n_rows * 100).round(pct_decimals),
            # Pandas dtype (int64, object, float64, …)
            "dtype": df.dtypes,
            # First observed non-null value — useful for a quick sanity check
            "Example": [
                df[col].dropna().iloc[0] if df[col].notna().any() else "-"
                for col in df.columns
            ],
        }
    ).rename_axis("Column")

    # Map the user-facing sort_by key to the actual column label
    sort_col = (
        "Missing"
        if sort_by == "missing"
        else "Sparsity %" if sort_by == "pct" else None
    )

    # Apply sort only when a valid sort column was resolved
    if sort_col:
        report = report.sort_values(sort_col, ascending=ascending)

    return report
```

```python
# ---------------------------------------------------------------------------
# Nested cross-validation utility
# ---------------------------------------------------------------------------
# Nested CV separates two concerns:
#   Outer loop — provides an *unbiased* estimate of the tuned model's
#                generalisation performance on held-out data.
#   Inner loop — selects the best hyperparameters via RandomizedSearchCV
#                on the training portion of each outer fold.
#
# Using GroupKFold in both loops ensures that all transactions belonging
# to the same account_id are always kept in the same fold, preventing
# account-level data leakage between train and validation splits.
# ---------------------------------------------------------------------------


@dataclass(frozen=True)
class NestedCVResult:
    """Immutable container for nested CV outputs.

    Attributes
    ----------
    outer_scores_pr_auc : NDArray[np.float64]
        PR-AUC (average precision) for each outer fold.
    outer_scores_roc_auc : NDArray[np.float64]
        ROC-AUC for each outer fold.
    best_params_per_fold : List[Dict[str, Any]]
        Best hyperparameter dict selected by the inner CV for each fold.
    """

    outer_scores_pr_auc: NDArray[np.float64]
    outer_scores_roc_auc: NDArray[np.float64]
    best_params_per_fold: List[Dict[str, Any]]


def run_nested_cv(
    *,
    X: Any,
    y: Any,
    groups: Optional[Any],
    pipeline: Pipeline,
    param_distributions: Dict[str, Any],
    outer_cv: Any,
    inner_cv: Any,
    n_iter: int,
    random_state: int,
    n_jobs: int,
) -> NestedCVResult:
    """Execute nested cross-validation with group-aware splits.

    Parameters
    ----------
    X : array-like or pd.DataFrame
        Feature matrix (including any columns needed for group extraction).
    y : array-like or pd.Series
        Binary target (0 = legitimate, 1 = fraud).
    groups : array-like or None
        Group labels for GroupKFold (typically ``account_id``).  Pass
        ``None`` to use standard stratified splitting.
    pipeline : sklearn Pipeline
        Preprocessing + model chain.  Step names must match the keys in
        *param_distributions* (e.g. ``"model__n_estimators"``).
    param_distributions : dict
        Hyperparameter grid passed to ``RandomizedSearchCV``.
    outer_cv : CV splitter
        Controls the outer evaluation loop (e.g. ``GroupKFold(n_splits=5)``).
    inner_cv : CV splitter
        Controls the inner tuning loop (e.g. ``GroupKFold(n_splits=4)``).
    n_iter : int
        Number of random hyperparameter combinations to evaluate in the
        inner loop per outer fold.
    random_state : int
        Seed for reproducibility of ``RandomizedSearchCV``.
    n_jobs : int
        Parallelism level (``-1`` = all available cores).

    Returns
    -------
    NestedCVResult
        Named tuple containing per-fold PR-AUC, ROC-AUC, and best params.
    """
    # Accumulate per-fold results
    pr_auc_scores: List[float] = []
    roc_auc_scores: List[float] = []
    best_params: List[Dict[str, Any]] = []

    # Build the split iterator; pass groups only when available
    split_iter = (
        outer_cv.split(X, y, groups=groups)
        if groups is not None
        else outer_cv.split(X, y)
    )

    for train_idx, test_idx in split_iter:
        # Slice training and held-out test subsets; support both
        # pandas DataFrames (iloc) and numpy arrays (direct indexing)
        X_train = X.iloc[train_idx] if hasattr(X, "iloc") else X[train_idx]
        y_train = y.iloc[train_idx] if hasattr(y, "iloc") else y[train_idx]
        X_test = X.iloc[test_idx] if hasattr(X, "iloc") else X[test_idx]
        y_test = y.iloc[test_idx] if hasattr(y, "iloc") else y[test_idx]

        # Slice groups to match the training indices for the inner CV
        groups_train = None
        if groups is not None:
            groups_train = (
                groups.iloc[train_idx] if hasattr(groups, "iloc") else groups[train_idx]
            )

        # --- Inner loop: randomised hyperparameter search ---------------
        # scoring="average_precision" optimises for PR-AUC, the preferred
        # metric for heavily imbalanced fraud detection tasks.
        search = RandomizedSearchCV(
            estimator=pipeline,
            param_distributions=param_distributions,
            n_iter=n_iter,
            scoring="average_precision",
            cv=inner_cv,
            refit=True,  # Refit best params on the full training fold
            random_state=random_state,
            n_jobs=n_jobs,
            verbose=0,
        )

        # Pass groups to the inner CV when using GroupKFold
        if groups_train is not None:
            search.fit(X_train, y_train, groups=groups_train)
        else:
            search.fit(X_train, y_train)

        # Extract the best-tuned pipeline and record its parameters
        best_model: BaseEstimator = search.best_estimator_
        best_params.append(search.best_params_)

        # --- Outer evaluation: score on the held-out test fold ----------
        # predict_proba returns [P(class=0), P(class=1)]; take column 1
        y_proba = best_model.predict_proba(X_test)[:, 1]

        pr_auc = float(average_precision_score(y_test, y_proba))
        pr_auc_scores.append(pr_auc)

        roc_auc = float(roc_auc_score(y_test, y_proba))
        roc_auc_scores.append(roc_auc)

    return NestedCVResult(
        outer_scores_pr_auc=np.array(pr_auc_scores, dtype=np.float64),
        outer_scores_roc_auc=np.array(roc_auc_scores, dtype=np.float64),
        best_params_per_fold=best_params,
    )
```

---

#### Velocity & Burst Behavior
- **`txn_velocity_ratio`** — `txn_count_1min / txn_count_5min` — captures acceleration within a burst window; a ratio near 1.0 signals a sustained high-frequency attack rather than a one-off spike
- **`amount_velocity_ratio`** — `transaction_amount / avg_txn_amount_30d` — how far this transaction deviates from the cardholder's spending norm (a continuous version of a rules-based spike flag)
- **`max_amount_ratio`** — `transaction_amount / max_txn_amount_90d` — flags whether this transaction exceeds the cardholder's historical ceiling

---

#### Geographic & Device Risk Compounding
- **`geo_device_risk_score`** — a multiplicative or additive compound of `cross_border_flag`, `is_new_device`, `device_country_mismatch`, and `is_new_merchant` — individual flags are weak; combined they represent a high-dimensional fraud posture
- **`country_pair`** — encoded combination of `merchant_country` + `cardholder_country` — certain country pairs have statistically elevated fraud rates and this captures directionality that `cross_border_flag` alone misses
- **`impossible_travel`** - boolean threshold feature to indicate unlikely distances between sequential POS or ATM channel transactions. 

---

#### Time-Gap & Behavioral Recency
- **`log_time_since_last_txn`** — `log1p(time_since_last_txn_sec)` — the raw seconds are heavily skewed; log-transforming creates a more model-friendly distribution and emphasizes the extreme short-gap cases (account takeover pattern)
- **`is_rapid_reuse`** — binary flag for `time_since_last_txn_sec < 60` AND `txn_count_1min > 1` — distinguishes true rapid reuse from coincidental timing

---

#### Card & Account Lifecycle Risk
- **`card_age_bucket`** — binned `card_age_days` into `[new: 0-30, young: 31-90, established: 91+]` — new cards have a non-linear fraud risk profile; a continuous age feature won't capture the sharp early-lifecycle risk curve
- **`account_card_age_ratio`** — `card_age_days / account_age_days` — a new card on an old account is different from a new card on a new account; this ratio captures card replacement behavior that may indicate prior compromise
- **`historical_x_anomaly`** — interaction of `historical_fraud_flag * anomaly_label` — a previously flagged cardholder triggering the anomaly detector is a qualitatively different signal than either alone

---

#### Merchant Risk Context
- **`merchant_risk_tier`** — binned `merchant_fraud_rate_30d` into quantile tiers — raw rates are noisy for low-volume merchants; tiering provides stability for tree-based models
- **`high_risk_mcc`** — binary flag for MCCs historically associated with CNP fraud (e.g., gambling, crypto, digital goods, gift cards) — `merchant_category_code` as a raw integer carries no ordinal meaning to a model

---

#### CVV & Authentication Signal
- **`cvv_fail_ratio`** — `cvv_fail_count_24h / (card_age_days + 1)` — normalizes failure count by card age; a card that has failed CVV twice in its first two days of existence is very different from an established card
- **`cvv_fail_x_new_device`** — `cvv_fail_count_24h * is_new_device` — CVV failures on a new device are a particularly strong account takeover signal

---

#### Anomaly Score Engineering
- **`anomaly_score_percentile`** — rank-transformed `anomaly_score` — models can be sensitive to the raw score distribution; a percentile rank is more robust across score distributions
- **`anomaly_x_amount_ratio`** — `anomaly_score * amount_velocity_ratio` — amplifies the anomaly signal when it co-occurs with an unusual spend amount

---


```python
# ---------------------------------------------------------------------------
# Data Cleansing, Imputation, and Addressing Missing Values
# ---------------------------------------------------------------------------
# Pipeline summary:
#   1. Drop pure-identifier columns (zero predictive signal)
#   2. Drop compromised leakage columns (anomaly_label, anomaly_score)
#   3. Drop degenerate columns (txn_day_of_week)
#   4. Audit missingness via the existing missing_report() utility
#   5. Impute numeric columns  → median  (robust to skew / outliers)
#   6. Impute categorical columns → mode  (most-frequent value)
#   7. Validate: assert zero NaNs remain before proceeding
#   8. Register all transformations in the df_registry provenance tracker
# ---------------------------------------------------------------------------

# ---------------------------------------------------------------------------
# Step 1 — Drop pure-identifier columns
# ---------------------------------------------------------------------------
# transaction_id, device_id, and merchant_id are surrogate keys that carry
# no intrinsic predictive signal for a supervised model.  They are dropped
# here but noted for potential rejoining during EDA (e.g. merchant-level
# fraud rate aggregation).  card_id is also a pure identifier; account_id
# is retained because account-level behavioural features may be engineered
# from it downstream.
ID_COLS_TO_DROP: list[str] = [
    "transaction_id",
    "device_id",
    "merchant_id",
    "card_id",
]

df_clean: pd.DataFrame = df_raw.drop(columns=ID_COLS_TO_DROP)

register(
    new_df=df_clean,
    new_name="df_clean",
    parent_name="df_raw",
    operation="drop_id_cols",
    options={"columns_dropped": ID_COLS_TO_DROP, "reason": "pure identifiers — zero signal"},
)

print(
    f"[drop_id_cols] {df_raw.shape[1]} → {df_clean.shape[1]} columns "
    f"({len(ID_COLS_TO_DROP)} removed)"
)


# ---------------------------------------------------------------------------
# Step 2 — Drop compromised leakage columns
# ---------------------------------------------------------------------------
# anomaly_label: 94.2% correlation with the target (is_fraud).
#   Retaining it would give the model a near-perfect shortcut that does
#   not exist at inference time — classic target leakage.
#
# anomaly_score: 60% correlation with is_fraud.
#   Softer leakage, but still problematic.  We will reconstruct an
#   analogous unsupervised anomaly score from first principles during the
#   modelling phase (e.g. Isolation Forest on the cleaned feature set).
LEAKAGE_COLS_TO_DROP: list[str] = [
    "anomaly_label",
    "anomaly_score",
]

df_clean = df_clean.drop(columns=LEAKAGE_COLS_TO_DROP)

register(
    new_df=df_clean,
    new_name="df_clean",
    parent_name="df_clean",
    operation="drop_leakage_cols",
    options={
        "columns_dropped": LEAKAGE_COLS_TO_DROP,
        "reason": (
            "target leakage — anomaly_label r=0.942 with is_fraud; "
            "anomaly_score r=0.60 with is_fraud"
        ),
    },
)

print(
    f"[drop_leakage_cols] Removed leakage columns: {LEAKAGE_COLS_TO_DROP}. "
    f"Shape now: {df_clean.shape}"
)


# ---------------------------------------------------------------------------
# Step 3 — Drop degenerate columns
# ---------------------------------------------------------------------------
# txn_day_of_week: despite representing a theoretically meaningful signal
# (day-of-week fraud patterns are real in production data), this column is
# degenerate in the current dataset due to a data generation artifact.
#
# Empirical evidence for dropping:
#   - Only 3 of 7 possible values are present (Mon=0, Tue=1, Sun=6).
#   - Tuesday has just 8 rows out of 60,000 (0.01% of data).
#   - Fraud rates across days are flat and indistinguishable from the
#     global base rate (Mon: 4.39%, Sun: 4.54%, overall: 4.47%).
#   - Chi-square test: χ²=1.08, p=0.583 — not significant.
#   - Point-biserial correlation with is_fraud: r=0.003, p=0.399.
#
# The derived cyclic features (txn_dow_sin, txn_dow_cos) will also be
# excluded from df_features since they are transformations of this column.
DEGENERATE_COLS_TO_DROP: list[str] = [
    "txn_day_of_week",
]

df_clean = df_clean.drop(columns=DEGENERATE_COLS_TO_DROP)

register(
    new_df=df_clean,
    new_name="df_clean",
    parent_name="df_clean",
    operation="drop_degenerate_cols",
    options={
        "columns_dropped": DEGENERATE_COLS_TO_DROP,
        "reason": (
            "degenerate — only 3 of 7 day values present; "
            "Tuesday n=8; chi2 p=0.583; r=0.003 with is_fraud"
        ),
    },
)

print(
    f"[drop_degenerate_cols] Removed degenerate columns: {DEGENERATE_COLS_TO_DROP}. "
    f"Shape now: {df_clean.shape}"
)


# ---------------------------------------------------------------------------
# Step 4 — Missingness audit (pre-imputation baseline)
# ---------------------------------------------------------------------------
# Run the existing missing_report() utility to surface any gaps before
# imputation.  This snapshot is retained for the write-up; post-imputation
# we will assert the report contains all zeros.
print("\n--- Missingness report (pre-imputation) ---")
pre_imputation_missing: pd.DataFrame = missing_report(df_clean)

# Surface only columns that actually have missing values to keep the output
# readable; if the dataset is fully complete, report that explicitly.
missing_cols: pd.DataFrame = pre_imputation_missing[
    pre_imputation_missing["Missing"] > 0
]

if missing_cols.empty:
    print("No missing values detected — imputation step will be a no-op.")
else:
    print(f"{len(missing_cols)} column(s) with missing values:\n")
    try:
        from IPython.display import display  # type: ignore[import-untyped]

        display(missing_cols)
    except ImportError:
        print(missing_cols.to_string())


# ---------------------------------------------------------------------------
# Steps 5 & 6 — Median imputation (numeric) and mode imputation (categorical)
# ---------------------------------------------------------------------------
# Imputation strategy rationale:
#   - Numeric → median: resistant to the heavy right-skew present in
#     transaction amounts, distances, and time deltas.  Mean imputation
#     would be pulled toward outlier fraud events.
#   - Categorical → mode: preserves the most-common observed category.
#     This is conservative — we are not inventing new category values.
#
# We operate column-by-column (rather than using sklearn's SimpleImputer
# inside a Pipeline here) to keep the logic transparent and auditable.
# The sklearn Pipeline handles imputation formally during model training.

# Identify column types after all drop steps have been applied
numeric_cols: list[str] = df_clean.select_dtypes(include=["number"]).columns.tolist()
categorical_cols: list[str] = df_clean.select_dtypes(
    include=["object", "category"]
).columns.tolist()

# Guard: remove the target from the numeric list — it must never be imputed
TARGET_COL: str = "is_fraud"
if TARGET_COL in numeric_cols:
    numeric_cols.remove(TARGET_COL)

# -- Numeric: fill with per-column median --
numeric_imputed_cols: list[str] = []

for col in numeric_cols:
    n_missing: int = int(df_clean[col].isna().sum())
    if n_missing > 0:
        col_median: float = df_clean[col].median()
        df_clean[col] = df_clean[col].fillna(col_median)
        numeric_imputed_cols.append(col)
        print(f"  [numeric impute] '{col}': {n_missing} NaN → median ({col_median:.4f})")

# -- Categorical: fill with per-column mode --
categorical_imputed_cols: list[str] = []

for col in categorical_cols:
    n_missing = int(df_clean[col].isna().sum())
    if n_missing > 0:
        col_mode: str = str(df_clean[col].mode().iloc[0])
        df_clean[col] = df_clean[col].fillna(col_mode)
        categorical_imputed_cols.append(col)
        print(f"  [categorical impute] '{col}': {n_missing} NaN → mode ('{col_mode}')")

register(
    new_df=df_clean,
    new_name="df_clean",
    parent_name="df_clean",
    operation="impute_missing_values",
    options={
        "numeric_strategy": "median",
        "categorical_strategy": "mode",
        "numeric_cols_imputed": numeric_imputed_cols,
        "categorical_cols_imputed": categorical_imputed_cols,
        "target_col_excluded": TARGET_COL,
    },
)


# ---------------------------------------------------------------------------
# Step 7 — Post-imputation validation
# ---------------------------------------------------------------------------
total_missing_post: int = int(df_clean.isna().sum().sum())

assert total_missing_post == 0, (
    f"Imputation incomplete — {total_missing_post} NaN(s) remain in df_clean. "
    "Inspect missing_report(df_clean) to identify the affected columns."
)

print(
    f"\n[validation] Post-imputation NaN count: {total_missing_post} ✓  "
    f"Final shape: {df_clean.shape}"
)


# ---------------------------------------------------------------------------
# Step 8 — Final shape summary
# ---------------------------------------------------------------------------
print(
    f"\n{'=' * 60}\n"
    f"Cleansing complete.\n"
    f"  Rows   : {df_clean.shape[0]:,}\n"
    f"  Columns: {df_clean.shape[1]:,}  "
    f"(from {df_raw.shape[1]} original; "
    f"{df_raw.shape[1] - df_clean.shape[1]} removed)\n"
    f"  Removed breakdown:\n"
    f"    Identifiers  : {len(ID_COLS_TO_DROP)}\n"
    f"    Leakage      : {len(LEAKAGE_COLS_TO_DROP)}\n"
    f"    Degenerate   : {len(DEGENERATE_COLS_TO_DROP)}\n"
    f"  Target : '{TARGET_COL}'  "
    f"(fraud rate = {df_clean[TARGET_COL].mean():.2%})\n"
    f"{'=' * 60}"
)

show_lineage(name_filter="df_clean", compact=True)
```

```python
# ---------------------------------------------------------------------------
# Register df_features as a versioned copy of df_clean, then append all
# engineered feature columns to it.
# ---------------------------------------------------------------------------
# We explicitly copy df_clean so that:
#   1. df_clean remains immutable — a clean rollback point at any stage
#   2. df_features has a registered provenance entry as its own lineage node
#   3. All feature construction writes are isolated to df_features
# ---------------------------------------------------------------------------

df_features: pd.DataFrame = df_clean.copy()

register(
    new_df=df_features,
    new_name="df_features",
    parent_name="df_clean",
    operation="copy_for_feature_engineering",
    options={"reason": "immutable base copy; all engineered columns appended here"},
)
# ---------------------------------------------------------------------------
# Feature Engineering — Engineered Features DataFrame
# ---------------------------------------------------------------------------
# This cell derives all engineered features from df_clean and assembles them
# into df_features.  No row is added or removed; df_features is a column-wise
# extension of df_clean with all new signal columns appended.
#
# Feature groups (in order of construction):
#   A. Cyclic time encoding          (txn_hour → sin/cos pair)
#   B. Velocity & burst ratios       (amount / count acceleration signals)
#   C. Geographic & device compounding (geo_device_risk_score, country_pair)
#   D. Impossible travel speed       (km/h; NaN-safe for web/online channels)
#   E. Time-gap recency features     (log-transform, rapid-reuse flag)
#   F. Card & account lifecycle      (age buckets, card/account ratio)
#   G. Merchant risk context         (risk tier, high-risk MCC flag)
#   H. CVV & authentication signals  (normalised fail ratio, interaction term)
#   I. Historical × anomaly compound (interaction of prior fraud + anomaly)
#   J. Normalised merchant fraud rate (smoothed; robust to low-volume noise)
#
# Black / ruff compliance notes:
#   - All lines ≤ 88 characters (black default)
#   - No bare `except`; all caught exceptions are typed
#   - No unused imports or variables
#   - Type annotations on all non-trivial intermediate scalars
# ---------------------------------------------------------------------------


# ===========================================================================
# A. Cyclic time encoding
# ===========================================================================
# txn_hour is a circular variable: hour 23 and hour 0 are adjacent, but a
# linear encoding treats them as maximally distant.  Projecting onto the unit
# circle (sin/cos pair) preserves the circular topology and is compatible with
# distance-based and gradient-based models alike.
#
# Formula: θ = 2π × (value / period)
#   txn_hour_sin = sin(2π × txn_hour / 24)
#   txn_hour_cos = cos(2π × txn_hour / 24)
#

# -- Hour of day --
_hour_theta: pd.Series = 2 * np.pi * df_features["txn_hour"] / 24
df_features["txn_hour_sin"] = np.sin(_hour_theta)
df_features["txn_hour_cos"] = np.cos(_hour_theta)

# NOTE: The day-of-week column is degenerate in this dataset. encoding removed. 
# -- Day of week --
#_dow_theta: pd.Series = 2 * np.pi * df_features["txn_day_of_week"] / 7
#df_features["txn_dow_sin"] = np.sin(_dow_theta)
#df_features["txn_dow_cos"] = np.cos(_dow_theta)

# The raw integer txn_hour column is retained in df_features so that tree-based
# models (which do not benefit from cyclic encoding) can also use it.
# The sklearn Pipeline will select the appropriate representation per model.

print(
    "[A] Cyclic time features added: "
    "txn_hour_sin, txn_hour_cos"
)


# ===========================================================================
# B. Velocity & burst ratio features
# ===========================================================================
# Raw velocity counts and amounts carry signal, but their ratio forms reveal
# the *shape* of activity (accelerating burst vs. steady-state) which the
# base columns cannot express individually.

# -- B1. Intra-window velocity acceleration --
# txn_count_1min / txn_count_5min: a ratio near 1.0 means all 5-minute
# activity is concentrated in the last minute — a characteristic burst
# signature of card-testing and credential-stuffing attacks.
# Add 1 to the denominator to avoid ZeroDivisionError on quiet windows.
df_features["txn_velocity_ratio"] = df_features["txn_count_1min"] / (
    df_features["txn_count_5min"] + 1
)

# -- B2. Amount relative to 30-day average (spend normalisation) --
# Answers: "How unusual is this amount for *this* cardholder?"
# A value of 5.0 means the transaction is 5× the cardholder's typical spend.
# +1 guard prevents division by zero for brand-new cards with no history.
df_features["amount_velocity_ratio"] = df_features["transaction_amount"] / (
    df_features["avg_txn_amount_30d"] + 1
)

# -- B3. Amount relative to 90-day maximum --
# Answers: "Does this transaction exceed the cardholder's historical ceiling?"
# Values > 1.0 flag transactions that are genuinely unprecedented for the card.
df_features["max_amount_ratio"] = df_features["transaction_amount"] / (
    df_features["max_txn_amount_90d"] + 1
)

print(
    "[B] Velocity & burst features added: "
    "txn_velocity_ratio, amount_velocity_ratio, max_amount_ratio"
)


# ===========================================================================
# C. Geographic & device risk compounding
# ===========================================================================
# Individual binary risk flags (cross_border, new_device, etc.) have limited
# discriminative power in isolation.  Their additive sum creates a risk
# posture score: 0 = no risk indicators, 4 = maximum simultaneous risk.

# -- C1. Additive geo-device risk score --
# Component flags (all are 0/1 integers in df_clean):
#   cross_border_flag       : merchant country ≠ cardholder country
#   is_new_device           : device not seen before for this account
#   device_country_mismatch : device IP country ≠ cardholder country
#   is_new_merchant         : merchant not seen before for this account
GEO_DEVICE_RISK_COMPONENTS: list[str] = [
    "cross_border_flag",
    "is_new_device",
    "device_country_mismatch",
    "is_new_merchant",
]
df_features["geo_device_risk_score"] = df_features[GEO_DEVICE_RISK_COMPONENTS].sum(
    axis=1
)

# -- C2. Country pair (merchant × cardholder) --
# Certain directed country pairs carry systematically elevated fraud risk that
# the symmetric cross_border_flag cannot capture (e.g. domestic card used
# at a specific foreign merchant country is riskier than the reverse).
# Encoded as a string concatenation; downstream one-hot or target-encoding
# will handle cardinality reduction.
df_features["country_pair"] = (
    df_features["merchant_country"] + "_" + df_features["cardholder_country"]
)

print(
    "[C] Geo & device features added: geo_device_risk_score, country_pair"
)


# ===========================================================================
# D. Impossible travel speed
# ===========================================================================
# For card-present transactions (POS / ATM), a cardholder cannot physically
# be in two places simultaneously.  Speed = distance / time gives a
# continuous measure of geographic implausibility.
#
# Exclusion logic:
#   - Online / web transactions: distance is not meaningful (cardholder is
#     not physically present at the merchant); set to NaN.
#   - time_since_last_txn_sec == 0: undefined speed; set to NaN to avoid
#     division-by-zero and spurious infinity values.
#
# Units: km/h  (distance_from_last_txn_km ÷ (seconds ÷ 3600))
#
# Interpretation guide (approximate thresholds for flagging):
#   > 900 km/h : faster than a commercial aircraft → virtually impossible
#   200–900    : requires air travel → suspicious for card-present
#   <  200     : plausible by car / train

_ONLINE_CHANNELS: set[str] = {"online", "web", "ecommerce"}

# Boolean mask: True for card-present rows only
_card_present_mask: pd.Series = ~df_features["channel"].str.lower().isin(
    _ONLINE_CHANNELS
)

# Boolean mask: True when time elapsed is strictly positive
_time_positive_mask: pd.Series = df_features["time_since_last_txn_sec"] > 0

# Initialise column with NaN; fill only where both masks are satisfied
df_features["travel_speed_kmh"] = np.nan

df_features.loc[
    _card_present_mask & _time_positive_mask, "travel_speed_kmh"
] = (
    df_features.loc[_card_present_mask & _time_positive_mask, "distance_from_last_txn_km"]
    / (
        df_features.loc[_card_present_mask & _time_positive_mask, "time_since_last_txn_sec"]
        / 3600
    )
)

# Binary flag for fast downstream rule application (threshold: 900 km/h)
# NaN rows (online transactions) evaluate as False — intentional.
_IMPOSSIBLE_SPEED_THRESHOLD_KMH: float = 900.0
df_features["impossible_travel_flag"] = (
    df_features["travel_speed_kmh"] > _IMPOSSIBLE_SPEED_THRESHOLD_KMH
).astype("Int8")  # nullable integer preserves NaN semantics

print(
    f"[D] Travel speed features added: travel_speed_kmh, impossible_travel_flag "
    f"(threshold: {_IMPOSSIBLE_SPEED_THRESHOLD_KMH} km/h; "
    f"online/zero-time rows → NaN)"
)


# ===========================================================================
# E. Time-gap recency features
# ===========================================================================
# -- E1. Log-transformed time since last transaction --
# time_since_last_txn_sec is heavily right-skewed (most transactions are
# seconds or minutes apart; occasional gaps are days long).  log1p(x) = log(x+1)
# compresses the long tail and amplifies the signal in the critical short-gap
# regime where account-takeover bursts occur.
df_features["log_time_since_last_txn"] = np.log1p(
    df_features["time_since_last_txn_sec"]
)

# -- E2. Rapid-reuse flag --
# Identifies the specific account-takeover pattern: multiple transactions
# within a 60-second window.  Both conditions must hold simultaneously;
# either alone is insufficient to flag genuine rapid reuse.
_RAPID_REUSE_TIME_THRESHOLD_SEC: int = 60
_RAPID_REUSE_COUNT_THRESHOLD: int = 1  # i.e. txn_count_1min > 1 means ≥ 2 txns

df_features["is_rapid_reuse"] = (
    (df_features["time_since_last_txn_sec"] < _RAPID_REUSE_TIME_THRESHOLD_SEC)
    & (df_features["txn_count_1min"] > _RAPID_REUSE_COUNT_THRESHOLD)
).astype(int)

print(
    "[E] Time-gap features added: log_time_since_last_txn, is_rapid_reuse "
    f"(time < {_RAPID_REUSE_TIME_THRESHOLD_SEC}s AND count_1min > "
    f"{_RAPID_REUSE_COUNT_THRESHOLD})"
)


# ===========================================================================
# F. Card & account lifecycle features
# ===========================================================================
# -- F1. Card age bucket (ordinal) --
# New cards exhibit a non-linear fraud risk curve: risk is highest in the
# first 30 days (issuance fraud, CNP testing) and drops sharply thereafter.
# A continuous card_age_days feature cannot capture this sharp step; buckets
# encode the qualitatively distinct lifecycle stages.
#
# Bucket definitions (based on industry-standard card lifecycle stages):
#   new        :   0–30 days  — issuance / activation window
#   young      :  31–90 days  — early-use, still elevated risk
#   established: 91–365 days  — normalised behaviour expected
#   mature     :    > 365 days — long-standing relationship
_CARD_AGE_BINS: list[int] = [0, 30, 90, 365, int(df_features["card_age_days"].max()) + 1]
_CARD_AGE_LABELS: list[str] = ["new", "young", "established", "mature"]

df_features["card_age_bucket"] = pd.cut(
    df_features["card_age_days"],
    bins=_CARD_AGE_BINS,
    labels=_CARD_AGE_LABELS,
    right=True,       # Intervals are (left, right]: (0,30], (30,90], …
    include_lowest=True,  # Ensures day-0 cards fall into 'new'
)

# -- F2. Card-to-account age ratio --
# A new card on an old account is qualitatively different from a new card on
# a new account.  The former may indicate a card replacement following a
# prior compromise; the latter is expected onboarding behaviour.
# +1 guard prevents ZeroDivisionError for day-0 accounts.
df_features["account_card_age_ratio"] = df_features["card_age_days"] / (
    df_features["account_age_days"] + 1
)

print(
    "[F] Lifecycle features added: card_age_bucket "
    f"(bins: {_CARD_AGE_LABELS}), account_card_age_ratio"
)


# ===========================================================================
# G. Merchant risk context
# ===========================================================================
# -- G1. Merchant fraud rate tier --
# Raw merchant_fraud_rate_30d values are noisy for low-volume merchants
# (a single fraud event on a new merchant inflates the rate to 1.0).
# Quantile-based tiering gives stable, model-friendly ordinal risk groups.
#
# Using 4 tiers (quartiles): low / medium / high / critical
_N_MERCHANT_TIERS: int = 4
_MERCHANT_TIER_LABELS: list[str] = ["low", "medium", "high", "critical"]

df_features["merchant_risk_tier"] = pd.qcut(
    df_features["merchant_fraud_rate_30d"],
    q=_N_MERCHANT_TIERS,
    labels=_MERCHANT_TIER_LABELS,
    duplicates="drop",  # Handles edge case where many rows share the same rate
)

# -- G2. High-risk MCC flag --
# Merchant Category Codes associated with card-not-present fraud, digital
# goods, and cash-equivalent products.  These MCCs are targeted because
# goods are non-reversible, instantly delivered, or easily liquidated.
#
# Reference: Visa/Mastercard fraud chargeback data + industry literature.
# MCCs included:
#   7995 — Betting / Gambling
#   6051 — Quasi-cash / Crypto exchanges
#   5816 — Digital goods / In-app purchases
#   5961 — Catalogue / Mail-order
#   6211 — Securities brokers (cash-equivalent)
#   5944 — Jewellery stores (high-value, liquid)
#   4829 — Money transfer / Wire
_HIGH_RISK_MCCS: set[int] = {7995, 6051, 5816, 5961, 6211, 5944, 4829}

df_features["high_risk_mcc"] = df_features["merchant_category_code"].isin(
    _HIGH_RISK_MCCS
).astype(int)

print(
    f"[G] Merchant features added: merchant_risk_tier "
    f"({_N_MERCHANT_TIERS} tiers), high_risk_mcc "
    f"({len(_HIGH_RISK_MCCS)} MCCs flagged)"
)


# ===========================================================================
# H. CVV & authentication signal features
# ===========================================================================
# -- H1. CVV failure rate normalised by card age --
# A card that has failed CVV twice on day 1 of its life is far more
# suspicious than an established card with the same raw count.  Normalising
# by card age converts an absolute count into a rate signal.
# +1 guard on card_age_days prevents ZeroDivisionError for same-day-issue cards.
df_features["cvv_fail_ratio"] = df_features["cvv_fail_count_24h"] / (
    df_features["card_age_days"] + 1
)

# -- H2. CVV failure × new device interaction --
# CVV failures on a known device may be typos; CVV failures on a *new* device
# indicate a third party attempting to authenticate a stolen card number.
# The multiplicative interaction amplifies only when both signals co-occur.
df_features["cvv_fail_x_new_device"] = (
    df_features["cvv_fail_count_24h"] * df_features["is_new_device"]
)

print(
    "[H] CVV features added: cvv_fail_ratio, cvv_fail_x_new_device"
)

# SECTION I NOTE: removed anomaly_label and anomaly_score columns from df_features to prevent leakage
# we will reconstruct an analogous unsupervised anomaly score from first principles during the modelling phase (e.g. Isolation Forest on the cleaned feature set).
#
# # ===========================================================================
# J. Normalised (smoothed) merchant fraud rate
# ===========================================================================
# Bayesian smoothing (additive / Laplace smoothing) shrinks per-merchant fraud
# rates toward the global prior for low-volume merchants.
#
# Formula:  smoothed_rate = (fraud_count + α) / (total_count + α / global_prior)
#
# Here we approximate using the observed rate and a pseudo-count (k=10),
# which is equivalent to assuming each merchant has at least 10 virtual
# observations at the global mean before any real data is observed.
# This prevents a single fraud event on a new merchant from producing a 1.0
# rate that dominates the signal.
#
# smoothed = (observed_rate × n_obs + prior × k) / (n_obs + k)
# Since n_obs is not directly available, we use merchant_fraud_rate_30d as the
# observed rate and estimate n_obs implicitly via a pseudo-count approach:
#   smoothed_rate = (raw_rate + k × global_prior) / (1 + k)

_SMOOTHING_PSEUDO_COUNT: float = 10.0  # Controls strength of shrinkage toward prior
_GLOBAL_FRAUD_PRIOR: float = float(df_features["merchant_fraud_rate_30d"].mean())

df_features["merchant_fraud_rate_smoothed"] = (
    df_features["merchant_fraud_rate_30d"] + _SMOOTHING_PSEUDO_COUNT * _GLOBAL_FRAUD_PRIOR
) / (1 + _SMOOTHING_PSEUDO_COUNT)

print(
    f"[J] Smoothed merchant fraud rate added: merchant_fraud_rate_smoothed "
    f"(pseudo-count k={_SMOOTHING_PSEUDO_COUNT:.0f}, "
    f"global prior={_GLOBAL_FRAUD_PRIOR:.4f})"
)


# ===========================================================================
# Final validation & registry
# ===========================================================================
# Identify all newly created feature columns (anything in df_features that
# was not in df_clean) for the provenance record.
new_feature_cols: list[str] = [
    col for col in df_features.columns if col not in df_clean.columns
]

# Sanity check: no NaNs introduced in columns where NaN is not intentional.
# travel_speed_kmh is excluded because NaN is by design for online channels.
_INTENTIONAL_NAN_COLS: set[str] = {"travel_speed_kmh", "impossible_travel_flag"}
_cols_to_nan_check = [c for c in new_feature_cols if c not in _INTENTIONAL_NAN_COLS]

_unexpected_nans: dict[str, int] = {
    col: int(df_features[col].isna().sum())
    for col in _cols_to_nan_check
    if df_features[col].isna().sum() > 0
}

if _unexpected_nans:
    raise ValueError(
        f"Unexpected NaNs found in engineered features: {_unexpected_nans}. "
        "Review the corresponding feature construction block above."
    )

# Register the full feature engineering step in the provenance tracker
register(
    new_df=df_features,
    new_name="df_features",
    parent_name="df_clean",
    operation="feature_engineering",
    options={
        "n_new_features": len(new_feature_cols),
        "new_columns": new_feature_cols,
        "groups": {
            "A_cyclic_time": ["txn_hour_sin", "txn_hour_cos", "txn_dow_sin", "txn_dow_cos"],
            "B_velocity": ["txn_velocity_ratio", "amount_velocity_ratio", "max_amount_ratio"],
            "C_geo_device": ["geo_device_risk_score", "country_pair"],
            "D_travel": ["travel_speed_kmh", "impossible_travel_flag"],
            "E_time_gap": ["log_time_since_last_txn", "is_rapid_reuse"],
            "F_lifecycle": ["card_age_bucket", "account_card_age_ratio"],
            "G_merchant": ["merchant_risk_tier", "high_risk_mcc"],
            "H_cvv": ["cvv_fail_ratio", "cvv_fail_x_new_device"],
            "I_interaction": ["historical_x_anomaly"],
            "J_smoothed_rate": ["merchant_fraud_rate_smoothed"],
        },
    },
)

# Summary output
print(
    f"\n{'=' * 60}\n"
    f"Feature engineering complete.\n"
    f"  Base columns   : {df_clean.shape[1]}\n"
    f"  New features   : {len(new_feature_cols)}\n"
    f"  Total columns  : {df_features.shape[1]}\n"
    f"  Rows           : {df_features.shape[0]:,}\n"
    f"  New feature list:\n"
    + "\n".join(f"    {i + 1:>2}. {col}" for i, col in enumerate(new_feature_cols))
    + f"\n{'=' * 60}"
)

show_lineage(name_filter="df_features", compact=True)
print(f"[init] df_features created from df_clean — shape: {df_features.shape}")
```

```python
# ---------------------------------------------------------------------------
# Nested cross-validation utility
# ---------------------------------------------------------------------------
# Nested CV separates two concerns:
#   Outer loop — provides an *unbiased* estimate of the tuned model's
#                generalisation performance on held-out data.
#   Inner loop — selects the best hyperparameters via RandomizedSearchCV
#                on the training portion of each outer fold.
#
# Using GroupKFold in both loops ensures that all transactions belonging
# to the same account_id are always kept in the same fold, preventing
# account-level data leakage between train and validation splits.
# ---------------------------------------------------------------------------


@dataclass(frozen=True)
class NestedCVResult:
    """Immutable container for nested CV outputs.

    Attributes
    ----------
    outer_scores_pr_auc : NDArray[np.float64]
        PR-AUC (average precision) for each outer fold.
    outer_scores_roc_auc : NDArray[np.float64]
        ROC-AUC for each outer fold.
    best_params_per_fold : List[Dict[str, Any]]
        Best hyperparameter dict selected by the inner CV for each fold.
    """

    outer_scores_pr_auc: NDArray[np.float64]
    outer_scores_roc_auc: NDArray[np.float64]
    best_params_per_fold: List[Dict[str, Any]]


def run_nested_cv(
    *,
    X: Any,
    y: Any,
    groups: Optional[Any],
    pipeline: Pipeline,
    param_distributions: Dict[str, Any],
    outer_cv: Any,
    inner_cv: Any,
    n_iter: int,
    random_state: int,
    n_jobs: int,
) -> NestedCVResult:
    """Execute nested cross-validation with group-aware splits.

    Parameters
    ----------
    X : array-like or pd.DataFrame
        Feature matrix (including any columns needed for group extraction).
    y : array-like or pd.Series
        Binary target (0 = legitimate, 1 = fraud).
    groups : array-like or None
        Group labels for GroupKFold (typically ``account_id``).  Pass
        ``None`` to use standard stratified splitting.
    pipeline : sklearn Pipeline
        Preprocessing + model chain.  Step names must match the keys in
        *param_distributions* (e.g. ``"model__n_estimators"``).
    param_distributions : dict
        Hyperparameter grid passed to ``RandomizedSearchCV``.
    outer_cv : CV splitter
        Controls the outer evaluation loop (e.g. ``GroupKFold(n_splits=5)``).
    inner_cv : CV splitter
        Controls the inner tuning loop (e.g. ``GroupKFold(n_splits=4)``).
    n_iter : int
        Number of random hyperparameter combinations to evaluate in the
        inner loop per outer fold.
    random_state : int
        Seed for reproducibility of ``RandomizedSearchCV``.
    n_jobs : int
        Parallelism level (``-1`` = all available cores).

    Returns
    -------
    NestedCVResult
        Named tuple containing per-fold PR-AUC, ROC-AUC, and best params.
    """
    # Accumulate per-fold results
    pr_auc_scores: List[float] = []
    roc_auc_scores: List[float] = []
    best_params: List[Dict[str, Any]] = []

    # Build the split iterator; pass groups only when available
    split_iter = (
        outer_cv.split(X, y, groups=groups)
        if groups is not None
        else outer_cv.split(X, y)
    )

    for train_idx, test_idx in split_iter:
        # Slice training and held-out test subsets; support both
        # pandas DataFrames (iloc) and numpy arrays (direct indexing)
        X_train = X.iloc[train_idx] if hasattr(X, "iloc") else X[train_idx]
        y_train = y.iloc[train_idx] if hasattr(y, "iloc") else y[train_idx]
        X_test = X.iloc[test_idx] if hasattr(X, "iloc") else X[test_idx]
        y_test = y.iloc[test_idx] if hasattr(y, "iloc") else y[test_idx]

        # Slice groups to match the training indices for the inner CV
        groups_train = None
        if groups is not None:
            groups_train = (
                groups.iloc[train_idx] if hasattr(groups, "iloc") else groups[train_idx]
            )

        # --- Inner loop: randomised hyperparameter search ---------------
        # scoring="average_precision" optimises for PR-AUC, the preferred
        # metric for heavily imbalanced fraud detection tasks.
        search = RandomizedSearchCV(
            estimator=pipeline,
            param_distributions=param_distributions,
            n_iter=n_iter,
            scoring="average_precision",
            cv=inner_cv,
            refit=True,  # Refit best params on the full training fold
            random_state=random_state,
            n_jobs=n_jobs,
            verbose=0,
        )

        # Pass groups to the inner CV when using GroupKFold
        if groups_train is not None:
            search.fit(X_train, y_train, groups=groups_train)
        else:
            search.fit(X_train, y_train)

        # Extract the best-tuned pipeline and record its parameters
        best_model: BaseEstimator = search.best_estimator_
        best_params.append(search.best_params_)

        # --- Outer evaluation: score on the held-out test fold ----------
        # predict_proba returns [P(class=0), P(class=1)]; take column 1
        y_proba = best_model.predict_proba(X_test)[:, 1]

        pr_auc = float(average_precision_score(y_test, y_proba))
        pr_auc_scores.append(pr_auc)

        roc_auc = float(roc_auc_score(y_test, y_proba))
        roc_auc_scores.append(roc_auc)

    return NestedCVResult(
        outer_scores_pr_auc=np.array(pr_auc_scores, dtype=np.float64),
        outer_scores_roc_auc=np.array(roc_auc_scores, dtype=np.float64),
        best_params_per_fold=best_params,
    )
```

```python
def ks_fraud_score(y_true, scores, plot=True, label="model"):
    """Compute the Kolmogorov-Smirnov statistic for a binary fraud classifier.

    The KS statistic measures the maximum vertical separation between the
    cumulative distribution functions (CDFs) of the fraud and legitimate
    score distributions:

        KS = max_t | F_fraud(t) - F_legit(t) |

    where:
        F_fraud(t) = P(score ≤ t | fraud)      — CDF of scores for fraud cases
        F_legit(t) = P(score ≤ t | legitimate) — CDF of scores for legit cases

    Intuitively, a perfect model would assign score = 1 to every fraud and
    score = 0 to every legitimate transaction, giving KS = 1.0.  A random
    model gives KS ≈ 0.0 because both CDFs rise at the same rate.

    In banking / credit-risk applications, KS is commonly used as the
    primary separation metric because it directly answers the operational
    question: "At the single best decision threshold, how well does the
    model separate the two populations?"  Unlike ROC-AUC, it gives you
    the threshold alongside the statistic.

    Rule of thumb for fraud models:
        KS < 0.20  — poor separation, model needs improvement
        0.20–0.40  — moderate separation
        0.40–0.60  — good separation
        > 0.60     — excellent separation (verify for overfitting)

    Parameters
    ----------
    y_true : array-like of int
        Ground-truth binary labels (0 = legitimate, 1 = fraud).
    scores : array-like of float
        Predicted fraud probabilities from model.predict_proba()[:, 1].
        Values must lie in [0, 1].
    plot : bool, optional
        If True (default), display the two CDFs with the KS gap annotated.
    label : str, optional
        Model name used in the plot title and printed summary.

    Returns
    -------
    ks_stat : float
        The KS statistic — maximum separation between the two CDFs.
        Lies in [0, 1]; higher is better.
    optimal_threshold : float
        The score threshold at which maximum separation occurs.
        Use this as a starting point for your decision threshold before
        further calibration against business cost constraints.

    Examples
    --------
    >>> ks, thresh = ks_fraud_score(y_test, model.predict_proba(X_test)[:, 1])
    >>> print(f"KS = {ks:.3f}, optimal threshold ≈ {thresh:.3f}")
    """
    y_true = np.asarray(y_true)
    scores = np.asarray(scores, dtype=float)

    if y_true.ndim != 1 or scores.ndim != 1:
        raise ValueError("y_true and scores must be 1-D arrays.")
    if len(y_true) != len(scores):
        raise ValueError(
            f"Length mismatch: y_true has {len(y_true)} elements "
            f"but scores has {len(scores)}."
        )
    if not np.isin(y_true, [0, 1]).all():
        raise ValueError("y_true must contain only 0 (legit) and 1 (fraud).")

    n_fraud = int(y_true.sum())
    n_legit = len(y_true) - n_fraud

    if n_fraud == 0 or n_legit == 0:
        raise ValueError(
            "y_true must contain at least one example of each class."
        )

    # -----------------------------------------------------------------------
    # Build the empirical CDFs
    # -----------------------------------------------------------------------
    # Sort every observation by ascending predicted score so we can sweep
    # the threshold from 0 → 1 in a single pass.
    order = np.argsort(scores)
    scores_sorted = scores[order]
    y_sorted = y_true[order]

    # At each score threshold t, the empirical CDF is:
    #   F_fraud(t) = (# fraud cases with score ≤ t) / n_fraud
    #   F_legit(t) = (# legit cases with score ≤ t) / n_legit
    #
    # cumsum over the sorted fraud indicators gives the running numerator
    # for F_fraud; (1 - y_sorted) gives the legit indicator.
    cdf_fraud = np.cumsum(y_sorted) / n_fraud
    cdf_legit = np.cumsum(1 - y_sorted) / n_legit

    # KS statistic = maximum absolute gap between the two CDFs
    gaps = np.abs(cdf_fraud - cdf_legit)
    ks_idx = int(np.argmax(gaps))
    ks_stat = float(gaps[ks_idx])
    optimal_threshold = float(scores_sorted[ks_idx])

    # -----------------------------------------------------------------------
    # Print summary
    # -----------------------------------------------------------------------
    print(f"--- KS Statistic: {label} ---")
    print(f"  KS statistic       : {ks_stat:.4f}")
    print(f"  Optimal threshold  : {optimal_threshold:.4f}")
    print(f"  Fraud cases        : {n_fraud:,}")
    print(f"  Legitimate cases   : {n_legit:,}")

    # -----------------------------------------------------------------------
    # Plot the two CDFs with the KS gap annotated
    # -----------------------------------------------------------------------
    if plot:
        fig, ax = plt.subplots(figsize=(9, 5))

        ax.plot(scores_sorted, cdf_fraud, color="#e74c3c", lw=2, label="Fraud CDF")
        ax.plot(scores_sorted, cdf_legit, color="#2980b9", lw=2, label="Legitimate CDF")

        # Vertical line at the optimal threshold
        ax.axvline(
            optimal_threshold,
            color="grey",
            linestyle="--",
            lw=1.2,
            label=f"Optimal threshold = {optimal_threshold:.3f}",
        )

        # Annotate the KS gap with a double-headed arrow
        y_fraud_at_ks = cdf_fraud[ks_idx]
        y_legit_at_ks = cdf_legit[ks_idx]
        ax.annotate(
            "",
            xy=(optimal_threshold, y_fraud_at_ks),
            xytext=(optimal_threshold, y_legit_at_ks),
            arrowprops=dict(arrowstyle="<->", color="black", lw=1.5),
        )
        ax.text(
            optimal_threshold + 0.01,
            (y_fraud_at_ks + y_legit_at_ks) / 2,
            f"KS = {ks_stat:.3f}",
            va="center",
            fontsize=10,
        )

        ax.set_xlabel("Predicted fraud score")
        ax.set_ylabel("Cumulative proportion")
        ax.set_title(f"KS Plot — {label}")
        ax.legend(loc="upper left")
        plt.tight_layout()
        plt.show()

    return ks_stat, optimal_threshold
```

```python
# ---------------------------------------------------------------------------
# LightGBM sklearn-compatible wrapper for run_nested_cv
# ---------------------------------------------------------------------------
# LGBMClassifier is already sklearn-compatible for predict/predict_proba,
# but early stopping requires eval_set at fit() time, which
# RandomizedSearchCV cannot provide directly.  This wrapper solves that by
# carving out an internal validation split from whatever training data it
# receives, using it solely for early stopping, then re-fitting on the full
# fold data at the best iteration before returning.
#
# This means run_nested_cv sees a standard sklearn estimator:
#   - fit(X, y)           ✓  (and optionally fit(X, y, groups=...))
#   - predict_proba(X)    ✓
#   - get_params()        ✓  (required by RandomizedSearchCV clone)
#   - set_params()        ✓  (required by RandomizedSearchCV clone)
# ---------------------------------------------------------------------------


class LGBMFraudClassifier(BaseEstimator):
    """Sklearn-compatible LightGBM wrapper with internal early stopping.

    Designed to be dropped into ``run_nested_cv`` as the estimator inside
    a sklearn ``Pipeline``.  Solves the early-stopping / eval_set problem
    by reserving a small internal validation split from the training data
    during ``fit()``, using it to find the best iteration, then re-fitting
    on the full training fold at that iteration count.

    Parameters
    ----------
    n_estimators : int
        Maximum boosting rounds before early stopping intervenes.
        Default: 1000.
    learning_rate : float
        Shrinkage per round.  Default: 0.05.
    num_leaves : int
        Maximum leaves per tree; controls model complexity.  Default: 63.
    min_child_samples : int
        Minimum samples per leaf; regularises against overfitting on rare
        fraud cases.  Default: 50.
    subsample : float
        Row subsampling fraction per round.  Default: 0.8.
    colsample_bytree : float
        Feature subsampling fraction per round.  Default: 0.8.
    reg_alpha : float
        L1 leaf-weight regularisation.  Default: 0.1.
    reg_lambda : float
        L2 leaf-weight regularisation.  Default: 1.0.
    early_stopping_rounds : int
        Stop if validation PR-AUC does not improve for this many rounds.
        Default: 50.
    early_stopping_val_size : float
        Fraction of training data to reserve for early stopping.  This
        split is internal to fit() and invisible to the outer CV loop.
        Default: 0.1 (10%).
    categorical_features : list[str] or None
        Column names to treat as categoricals natively.  Default: None.
    random_state : int
        Reproducibility seed.  Default: 7.

    Notes
    -----
    ``is_unbalance=True`` is always applied; it is not exposed as a
    parameter because it should always be set for imbalanced fraud data.

    The ``BaseEstimator`` mixin provides ``get_params()`` and
    ``set_params()`` automatically from ``__init__`` argument names,
    which is required for ``RandomizedSearchCV`` to clone and configure
    the estimator correctly across folds.
    """

    def __init__(
        self,
        n_estimators: int = 1000,
        learning_rate: float = 0.05,
        num_leaves: int = 63,
        min_child_samples: int = 50,
        subsample: float = 0.8,
        colsample_bytree: float = 0.8,
        reg_alpha: float = 0.1,
        reg_lambda: float = 1.0,
        early_stopping_rounds: int = 50,
        early_stopping_val_size: float = 0.1,
        categorical_features: list[str] | None = None,
        random_state: int = 7,
    ) -> None:
        # NOTE: parameter names here must exactly match the attribute names
        # assigned below.  BaseEstimator.get_params() inspects __init__
        # signatures via introspection — any mismatch breaks cloning.
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.num_leaves = num_leaves
        self.min_child_samples = min_child_samples
        self.subsample = subsample
        self.colsample_bytree = colsample_bytree
        self.reg_alpha = reg_alpha
        self.reg_lambda = reg_lambda
        self.early_stopping_rounds = early_stopping_rounds
        self.early_stopping_val_size = early_stopping_val_size
        self.categorical_features = categorical_features
        self.random_state = random_state

        # Populated during fit(); None before then
        self.model_: Any = None
        self.best_iteration_: int | None = None

    def fit(
        self,
        X: pd.DataFrame,
        y: pd.Series,
        **fit_params: Any,  # Absorbs groups= from run_nested_cv without error
    ) -> "LGBMFraudClassifier":
        """Fit LightGBM with internal early stopping.

        Splits *X* / *y* into a fit portion and an internal validation
        portion.  Uses the validation portion to find the best iteration
        via early stopping, then re-fits on the full *X* / *y* at that
        fixed iteration count so the final model sees all training data.

        Parameters
        ----------
        X : pd.DataFrame
            Training feature matrix for this fold.
        y : pd.Series
            Binary training labels.
        **fit_params
            Silently absorbs keyword arguments forwarded by sklearn
            (e.g. ``groups=`` from ``GroupKFold``).

        Returns
        -------
        self
        """
        try:
            import lightgbm as lgb  # type: ignore[import-untyped]
        except ImportError as exc:
            raise ImportError(
                "LightGBM is not installed.  Run: pip install lightgbm --quiet"
            ) from exc

        # -------------------------------------------------------------------
        # Step 1: internal train/validation split for early stopping
        # -------------------------------------------------------------------
        # stratify=y preserves the fraud rate in both splits, which is
        # critical given the ~4.5% minority class.
        X_fit, X_es, y_fit, y_es = train_test_split(
            X,
            y,
            test_size=self.early_stopping_val_size,
            stratify=y,
            random_state=self.random_state,
        )

        # -------------------------------------------------------------------
        # Step 2: fit with early stopping to find the best iteration
        # -------------------------------------------------------------------
        _probe = lgb.LGBMClassifier(**self._lgbm_params())

        _probe.fit(
            X_fit,
            y_fit,
            eval_set=[(X_es, y_es)],
            eval_names=["early_stop_val"],
            categorical_feature=self.categorical_features or "auto",
            callbacks=[
                lgb.early_stopping(
                    stopping_rounds=self.early_stopping_rounds,
                    first_metric_only=True,
                    verbose=False,  # Silence per-fold chatter inside CV
                ),
                lgb.log_evaluation(period=0),  # Fully suppress training log
            ],
        )

        # Record the best iteration found on the internal validation set
        self.best_iteration_ = int(_probe.best_iteration_)

        # -------------------------------------------------------------------
        # Step 3: re-fit on the full fold at the best fixed iteration count
        # -------------------------------------------------------------------
        # This ensures the final model benefits from 100% of the fold's
        # training data, not just the 90% used in the early stopping probe.
        # n_estimators is fixed at best_iteration_; no eval_set needed.
        self.model_ = lgb.LGBMClassifier(
            **{**self._lgbm_params(), "n_estimators": self.best_iteration_}
        )

        self.model_.fit(
            X,
            y,
            categorical_feature=self.categorical_features or "auto",
            callbacks=[lgb.log_evaluation(period=0)],
        )

        return self

    def predict_proba(self, X: pd.DataFrame) -> np.ndarray:
        """Return class probability estimates.

        Parameters
        ----------
        X : pd.DataFrame
            Feature matrix to score.

        Returns
        -------
        np.ndarray of shape (n_samples, 2)
            Columns are [P(legitimate), P(fraud)].
        """
        self._check_is_fitted()
        return self.model_.predict_proba(X)  # type: ignore[union-attr]

    def predict(self, X: pd.DataFrame) -> np.ndarray:
        """Return hard class predictions at the 0.5 threshold.

        Parameters
        ----------
        X : pd.DataFrame
            Feature matrix to score.

        Returns
        -------
        np.ndarray of shape (n_samples,)
            Binary predictions (0 = legitimate, 1 = fraud).
        """
        self._check_is_fitted()
        return self.model_.predict(X)  # type: ignore[union-attr]

    # -----------------------------------------------------------------------
    # Internal helpers
    # -----------------------------------------------------------------------

    def _lgbm_params(self) -> dict[str, Any]:
        """Assemble the LGBMClassifier constructor kwargs from self.

        Centralised so both the early-stopping probe and the final
        re-fit use identical parameters (except n_estimators).

        Returns
        -------
        dict[str, Any]
            Keyword arguments for ``lgb.LGBMClassifier()``.
        """
        return {
            "n_estimators": self.n_estimators,
            "learning_rate": self.learning_rate,
            "num_leaves": self.num_leaves,
            "min_child_samples": self.min_child_samples,
            "subsample": self.subsample,
            "subsample_freq": 1,
            "colsample_bytree": self.colsample_bytree,
            "reg_alpha": self.reg_alpha,
            "reg_lambda": self.reg_lambda,
            "is_unbalance": True,
            "objective": "binary",
            "metric": "average_precision",
            "random_state": self.random_state,
            "n_jobs": -1,
            "verbose": -1,
        }

    def _check_is_fitted(self) -> None:
        """Raise if predict is called before fit.

        Raises
        ------
        RuntimeError
            If ``fit()`` has not been called yet.
        """
        if self.model_ is None:
            raise RuntimeError(
                "This LGBMFraudClassifier instance is not fitted yet. "
                "Call fit() before predict() or predict_proba()."
            )
```

```python
def histogram(col, title, xlabel, ylabel):
    """Plot a histogram with a KDE overlay for a single feature.

    Parameters
    ----------
    col : array-like
        The data to plot (a pandas Series or numpy array).
    title : str
        Chart title displayed above the plot.
    xlabel : str
        Label for the x-axis (feature name / units).
    ylabel : str
        Label for the y-axis (typically 'Count' or 'Density').
    """
    # Wide, short figure keeps the histogram readable without wasting
    # vertical space on a simple univariate distribution.
    plt.figure(figsize=(10, 4))

    # 50 bins gives enough granularity for most continuous features;
    # kde=True overlays a smoothed density estimate for shape interpretation.
    sns.histplot(col, bins=50, kde=True)

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
def barchart(
    title,
    xlabel,
    ylabel,
    stat="count",
    x=None,
    y=None,
    hue=None,
    data=None,
    size=(10, 5),
):
    """Plot a count / frequency bar chart for a categorical feature.

    Parameters
    ----------
    title : str
        Chart title.
    xlabel : str
        Label for the x-axis.
    ylabel : str
        Label for the y-axis.
    stat : str, optional
        Seaborn countplot statistic — ``'count'`` (default) or
        ``'percent'``.
    x : str or None
        Column name to plot on the x-axis.
    y : str or None
        Column name to plot on the y-axis (for horizontal bars).
    hue : str or None
        Column name used to colour bars (e.g. ``'is_fraud'``).
    data : pd.DataFrame
        Source DataFrame containing *x*, *y*, and *hue* columns.
    size : tuple of (float, float), optional
        Figure size in inches as ``(width, height)``.
    """
    plt.figure(figsize=size)

    # countplot counts observations per category; passing stat allows
    # switching to normalised proportions when raw counts are misleading.
    sns.countplot(data=data, stat=stat, x=x, y=y, hue=hue)

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    # Rotate long category labels to prevent overlap
    plt.xticks(rotation=45, ha="right")

    # Show the hue legend only when a grouping variable was provided
    if hue is not None:
        plt.legend(title=hue)

    plt.tight_layout()
    plt.show()
```

```python
def boxplot(x, y, data, title, xlabel, ylabel):
    """Plot a grouped boxplot to compare distributions across categories.

    Parameters
    ----------
    x : str
        Column name for the categorical grouping variable (x-axis).
    y : str
        Column name for the numeric variable whose distribution is shown.
    data : pd.DataFrame
        Source DataFrame.
    title : str
        Chart title.
    xlabel : str
        Label for the x-axis.
    ylabel : str
        Label for the y-axis.
    """
    plt.figure(figsize=(10, 5))

    # Seaborn boxplot: box spans the IQR, whiskers extend to 1.5× IQR,
    # and individual points beyond that are plotted as outliers.
    sns.boxplot(x=x, y=y, data=data)

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    # Rotate category labels so they do not overlap when there are many
    # categories (e.g. merchant_category_code).
    plt.xticks(rotation=45, ha="right")

    plt.tight_layout()
    plt.show()
```

```python
def stacked_barchart(x, y, xlabel, ylabel, title):
    """Plot a stacked proportional bar chart for a binary target by category.

    Each bar represents one category of *x* and shows the proportion of
    each *y* class within that category.  This is useful for visualising
    how the fraud rate varies across categorical features such as channel
    or entry_mode.

    Parameters
    ----------
    x : array-like
        Categorical grouping variable (e.g. ``df['channel']``).
    y : array-like
        Binary target variable (e.g. ``df['is_fraud']``).
    xlabel : str
        Label for the x-axis.
    ylabel : str
        Label for the y-axis (typically 'Proportion').
    title : str
        Chart title.
    """
    # Combine x and y into a temporary DataFrame for groupby operations
    df = pd.DataFrame({"x": x, "y": y})

    # Compute the within-group normalised proportions of each y class.
    # unstack() pivots y categories into columns; fillna(0) handles
    # any category / class combinations that do not appear in the data.
    # The result is a DataFrame where rows are values of 'x' and
    # columns are categories of 'y'.
    prop_df = df.groupby("x")["y"].value_counts(normalize=True).unstack().fillna(0)

    # Two-colour palette: blue for class 0, orange for class 1
    colors = ["#1f77b4", "#ff7f0e"]  # Adjust for more than two classes

    # Scale figure width with the number of categories; 8 inches minimum
    num_categories = prop_df.shape[0]
    width = max(8, num_categories * 0.6)

    # Stacked bars with normalised proportions (each bar totals 1.0)
    prop_df.plot(kind="bar", stacked=True, figsize=(width, 5), color=colors)

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(title="y")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
def pairplot_with_hue(data, vars_to_plot, hue_col, title=""):
    """Plot a pairwise scatter matrix coloured by a binary target column.

    Diagonal cells show per-class KDE curves; off-diagonal cells show
    scatter plots for every combination of variables.  Useful for quickly
    spotting which feature pairs best separate fraud from legitimate
    transactions.

    Parameters
    ----------
    data : pd.DataFrame
        Source DataFrame containing all relevant columns.
    vars_to_plot : list of str
        Feature columns to include in the pairplot axes.
    hue_col : str
        Binary column used to colour points (e.g. ``'is_fraud'``).
    title : str, optional
        Super-title placed above the entire figure.
    """
    # diag_kind="kde" shows per-class density curves on the diagonal,
    # which is more informative than overlapping histograms for two classes.
    sns.pairplot(
        data[vars_to_plot + [hue_col]],
        hue=hue_col,
        palette={0: "#1f77b4", 1: "#ff7f0e"},
        diag_kind="kde",
    )

    # y=1.02 lifts the title above the top row of subplots
    plt.suptitle(title, y=1.02)
    plt.show()
```

Identifier columns such as transaction IDs, account IDs, card IDs, merchant IDs, device IDs, and timestamps are unique or near-unique values. These columns describe how and where a transaction occurred, such as the transaction channel, card entry method, merchant type, geographic information, and CVV verification outcome.

Machine learning models cannot directly interpret categorical values in their raw form. Therefore, these columns are explicitly grouped so they can be encoded into numerical representations (e.g., one-hot encoding or target encoding) during preprocessing

```python
def pr_plot(y_true, scores, label="model"):
    """Plot a Precision-Recall curve and return the Average Precision score.

    PR-AUC (Average Precision) is the recommended primary metric for
    imbalanced fraud detection because it is not inflated by the large
    number of true negatives — unlike ROC-AUC.

    Parameters
    ----------
    y_true : array-like of int
        Ground-truth binary labels (0 = legitimate, 1 = fraud).
    scores : array-like of float
        Predicted fraud probabilities from ``model.predict_proba()[:, 1]``.
    label : str, optional
        Legend label; the AP score is appended automatically.

    Returns
    -------
    float
        Average Precision (area under the PR curve).
    """
    # Compute the full precision-recall curve across all decision thresholds
    p, r, _ = precision_recall_curve(y_true, scores)

    # Scalar summary of the curve area (equivalent to AP)
    ap = average_precision_score(y_true, scores)

    plt.figure()
    plt.plot(r, p, lw=2, label=f"{label} (AP={ap:.3f})")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(f"PR Curve - {label}")
    plt.legend()
    plt.show()

    return ap
```

```python
def best_f1_threshold(y_true, scores):
    """Find the decision threshold that maximises the F1 score.

    The default 0.5 threshold is rarely optimal for imbalanced datasets.
    This function sweeps all thresholds returned by the PR curve and
    selects the one that yields the highest F1 score.

    Parameters
    ----------
    y_true : array-like of int
        Ground-truth binary labels (0 = legitimate, 1 = fraud).
    scores : array-like of float
        Predicted fraud probabilities from ``model.predict_proba()[:, 1]``.

    Returns
    -------
    tuple of (float, float, float, float)
        ``(threshold, f1, precision, recall)`` at the optimal threshold.
    """
    # precision_recall_curve returns arrays of length n_thresholds+1;
    # the last element of p and r corresponds to threshold → ∞ (precision=1,
    # recall=0) and has no matching threshold, so we exclude it with [:-1].
    p, r, th = precision_recall_curve(y_true, scores)

    # F1 = 2 * P * R / (P + R); add 1e-9 to avoid division by zero when
    # both precision and recall are zero at extreme thresholds.
    f1 = 2 * p[:-1] * r[:-1] / (p[:-1] + r[:-1] + 1e-9)

    # Index of the threshold that maximises F1 (ignoring NaN values)
    i = np.nanargmax(f1)

    return float(th[i]), float(f1[i]), float(p[i]), float(r[i])
```

```python
def evaluate_binary(y_true, y_pred, name):
    """Print a confusion matrix and full classification report.

    Parameters
    ----------
    y_true : array-like of int
        Ground-truth binary labels (0 = legitimate, 1 = fraud).
    y_pred : array-like of int
        Hard predicted labels (after applying a decision threshold).
    name : str
        Descriptive model name printed as a section header.
    """
    print(f"--- {name} ---")

    # Confusion matrix: rows = actual, cols = predicted
    # [[TN, FP],
    #  [FN, TP]]
    print("Confusion matrix:\n", confusion_matrix(y_true, y_pred))

    # Per-class precision, recall, F1, and support; digits=3 for readability
    print(classification_report(y_true, y_pred, digits=3))
```

Before moving ahead, we define a function to check the performance of the model using different metrics.

- We will be using metric functions defined in sklearn for accuracy, precision, recall, f1_score
- We will create a function which will print out all the above metrics in one go.

```python
def model_performance_classification(model, predictors, target):
    """Compute a concise classification performance summary for a fitted model.

    Generates hard predictions using the model's default 0.5 threshold
    and returns precision, recall, F1, and accuracy in a single-row
    DataFrame for easy comparison across models.

    Note: For fraud detection, prefer ``average_precision_score`` (PR-AUC)
    as the primary metric.  The metrics here use the default 0.5 threshold
    which is typically suboptimal on imbalanced data.

    Parameters
    ----------
    model : fitted sklearn estimator
        A model that implements ``predict()``.
    predictors : array-like or pd.DataFrame
        Feature matrix for which predictions are generated.
    target : array-like or pd.Series
        True binary labels (0 = legitimate, 1 = fraud).

    Returns
    -------
    pd.DataFrame
        Single-row DataFrame with columns:
        ``Accuracy``, ``Recall``, ``Precision``, ``F1``.
    """
    # Generate hard class predictions (0 or 1) at the default 0.5 threshold
    pred = model.predict(predictors)

    # Compute each metric independently so they can be inspected separately
    acc = accuracy_score(target, pred)  # Fraction of correct predictions
    recall = recall_score(target, pred)  # Fraud capture rate (TP / (TP+FN))
    precision = precision_score(target, pred)  # Alert quality (TP / (TP+FP))
    f1 = f1_score(target, pred)  # Harmonic mean of precision and recall

    # Return as a single-row DataFrame for easy concatenation / comparison
    df_perf = pd.DataFrame(
        {
            "Accuracy": acc,
            "Recall": recall,
            "Precision": precision,
            "F1": f1,
        },
        index=[0],
    )

    return df_perf
```

```python
def plot_confusion_matrix(model, predictors, target):
    """Plot a confusion matrix heatmap with raw counts and percentages.

    Each cell shows the raw count and the percentage of all predictions,
    making it easy to compare models at a glance regardless of dataset size.

    Parameters
    ----------
    model : fitted sklearn estimator
        A model that implements ``predict()``.
    predictors : array-like or pd.DataFrame
        Feature matrix used to generate predictions.
    target : array-like or pd.Series
        True binary labels (0 = legitimate, 1 = fraud).
    """
    # Generate hard predictions at the model's default threshold
    y_pred = model.predict(predictors)

    # Build the 2×2 confusion matrix
    cm = confusion_matrix(target, y_pred)

    # Construct annotation labels: raw count on line 1, percentage on line 2.
    # cm.flatten() gives [TN, FP, FN, TP]; reshape back to (2, 2) for the
    # heatmap annotation grid.
    labels = np.asarray(
        [
            ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
            for item in cm.flatten()
        ]
    ).reshape(2, 2)

    plt.figure(figsize=(6, 4))

    # fmt="" because our annotation strings already contain newlines and
    # special characters that seaborn's numeric formatter cannot handle.
    sns.heatmap(cm, annot=labels, fmt="")

    plt.ylabel("True label")  # Rows: actual class
    plt.xlabel("Predicted label")  # Columns: predicted class
    plt.tight_layout()
    plt.show()
```

Approach Overview
This is a well-structured fraud detection problem with rich behavioral, contextual, and risk signal data. Here’s how I’d think through the full analytical pipeline.

EDA Strategy
The first priority is understanding the class imbalance — fraud datasets are typically 0.1–2% positive, and that reality shapes nearly every subsequent decision. You’ll want to quantify this immediately and keep it in mind when interpreting any summary statistics.
Univariate analysis should cover distributions of transaction amounts (expect heavy right-skew — log-transform for visualization), temporal patterns (txn_hour, txn_day_of_week), card/account age distributions, and frequency counts for all categoricals. Crucially, do this separately for fraud vs. non-fraud cases to surface the most intuitive signals early — you’ll likely see that fraud clusters at certain hours, channels, and CVV failure rates immediately.
Bivariate and behavioral analysis is where this dataset gets interesting. Key relationships worth deep exploration:
	∙	Transaction amount vs. avg_txn_amount_30d (how much is this transaction relative to normal for this card?) — compute a ratio feature here
	∙	txn_count_1min and txn_count_5min vs. fraud rate — velocity spikes are one of the strongest fraud signals
	∙	distance_from_last_txn_km relative to time_since_last_txn_sec — you can derive an implied travel speed, which is a powerful impossible-travel indicator
	∙	merchant_fraud_rate_30d distributions split by is_fraud
	∙	Cross-border flag + device_country_mismatch + cardholder_country combinations — these stack risk signals in interesting ways
Correlation and multicollinearity — use a Spearman matrix (not Pearson) since many variables are skewed or ordinal. Watch for high correlation between txn_count_1min and txn_count_5min, and between the various amount fields. You don’t necessarily want to drop correlated features in a tree-based model, but understanding the structure helps with interpretation later.
Anomaly score analysis deserves its own section. Since you already have anomaly_score and anomaly_label, you should analyze how well these pre-computed signals align with is_fraud — are all anomalies fraudulent? Are there fraudulent transactions with low anomaly scores? A confusion matrix between anomaly_label and is_fraud, plus a distribution plot of anomaly_score stratified by is_fraud, will tell you how much independent signal this adds versus being redundant.

Feature Engineering Worth Prioritizing
Beyond the raw variables, these derived features tend to add meaningful lift:
	∙	Amount deviation ratio: transaction_amount / avg_txn_amount_30d — captures “unusually large for this card”
	∙	Amount ceiling breach: transaction_amount / max_txn_amount_90d — flags transactions exceeding historical maximum
	∙	Impossible travel speed: distance_from_last_txn_km / (time_since_last_txn_sec / 3600) — if speed > ~900 km/h it’s physically impossible; flag these
	∙	Risk stacking score: sum of boolean risk indicators (cross_border_flag, is_new_merchant, is_new_device, device_country_mismatch, historical_fraud_flag) — a simple count of how many risk flags are active simultaneously
	∙	CVV risk: combine cvv_result (failed) with cvv_fail_count_24h into a composite signal
	∙	Velocity ratio: txn_count_1min relative to the card’s typical behavior (though you may not have a per-card baseline readily available)
	∙	Time-of-day risk bins: group txn_hour into business hours, evening, late night — fraud patterns differ meaningfully
For categoricals like channel, entry_mode, cvv_result, and merchant_category_code, target encoding (mean fraud rate per category) often outperforms one-hot encoding in tree models and avoids high cardinality issues. Be careful to do this within cross-validation folds to prevent leakage.

A few specific analyses that would strengthen the notebook’s analytical rigor:

	∙	Chi-square tests for categorical features vs. is_fraud

Modeling Approach
Random Forest is a solid and defensible starting point for this problem — it handles class imbalance reasonably well with class_weight=‘balanced’, requires minimal preprocessing, is naturally resistant to outliers, and provides feature importances. But I’d recommend going further.
Gradient Boosting (XGBoost or LightGBM) should be your primary model. For fraud detection specifically, it tends to outperform Random Forest because it’s better at capturing subtle, non-linear interactions between features like the combination of a new device + cross-border + high CVV failures + unusual amount. LightGBM in particular handles the class imbalance well via scale_pos_weight, trains quickly on large datasets, and natively handles categoricals. Given that your business questions explicitly ask about hyperparameter tuning (Q4) and feature importance (Q5), XGBoost/LightGBM gives you a richer tuning surface and SHAP integration is first-class.
A reasonable modeling progression for the notebook:

	2.	Random Forest — answers your Q3 framing and provides a strong benchmark
	3.	XGBoost or LightGBM — likely your best performer and most tunable
	4.	(Optional) An Isolation Forest or Autoencoder as an unsupervised complement, given you already have anomaly scores in the data — interesting to compare against
Class imbalance handling — test SMOTE (on training data only, never on validation), class weighting, and threshold optimization separately. In fraud detection, optimizing for the right threshold on predicted probability is often more impactful than the choice of resampling technique. Your evaluation metric should be PR-AUC (Average Precision) rather than ROC-AUC — with severe imbalance, ROC-AUC is misleadingly optimistic.
For Q3 (real-time assessment), focus the notebook section on threshold analysis: plot precision-recall tradeoff at different operating thresholds, simulate what happens at a threshold tuned for 90% recall versus 95% precision, and discuss the business cost asymmetry between false positives (customer friction) and false negatives (fraud losses).

Key parameters to tune in XGBoost: n_estimators, max_depth, learning_rate, subsample, colsample_bytree, and scale_pos_weight.
For Q6 (real-time action rules), the combination of predicted fraud probability + anomaly_score naturally supports a tiered decision framework — for example, high fraud probability + high anomaly score → block; high fraud probability + low anomaly score → step-up authentication; low fraud probability + high anomaly score → flag for review.
